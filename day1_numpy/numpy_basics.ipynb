# broadcasting

```python
import numpy as np

x = np.array([12,34,34,2,43,76,82])
y = x+210
print(y)

mean = x.mean(axis=0)
print(mean)
z = x-mean
print(z)
```

# dot product

```python
a = np.array([1,2,3,4,5])
b = np.array([[2,23,43,54,87],
            [23,5,54,67,45],
            [23,54,87,2,82]])
print(b@a)

c = np.array([[12,34,56],
            [23,5,54],
            [23,54,87],
            [23,54,87],
            [90,82,55]
            ])
print(b@c)
```

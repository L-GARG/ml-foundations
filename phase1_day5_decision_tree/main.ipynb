```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine

data = load_wine()

x = data.data
y = data.target

print(x.shape)
print(np.unique(y))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42, shuffle = True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

tree2 = DecisionTreeClassifier(max_depth=2, random_state=42)
tree2.fit(x_train, y_train)

y_pred2 = tree2.predict(x_test)
acc2 = accuracy_score(y_test, y_pred2)
print(acc2)
```

```python
tree5 = DecisionTreeClassifier(max_depth=5, random_state=42)
tree5.fit(x_train, y_train)

y_pred5 = tree5.predict(x_test)
acc5 = accuracy_score(y_test, y_pred5)
print(acc5)
```

```python
tree10 = DecisionTreeClassifier(max_depth=10, random_state=42)
tree10.fit(x_train, y_train)

y_pred10 = tree10.predict(x_test)
acc10 = accuracy_score(y_test, y_pred10)
print(acc10)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(15,10))
plot_tree(tree2, feature_names=data.feature_names, class_names=data.target_names, filled=True)
plt.show()
```
